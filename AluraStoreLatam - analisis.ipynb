### Importación de datos



```python
import pandas as pd
import altair as alt
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = [f"Tienda {i+1}" for i in range(len(tiendas))]

facturacion_real = [df['Precio'].sum() for df in tiendas]

df_facturacion = pd.DataFrame({
   'Tienda': nombres_tiendas,
    'Facturación Numerica': facturacion_real,
    'Facturación COP': [f"${x:,.2f}" for x in facturacion_real]
}).sort_values(by='Facturación Numerica', ascending=False)

print(df_facturacion[['Tienda', 'Facturación COP']].to_string(index=False))
```

```python
import plotly.express as px


fig = px.bar(
    df_facturacion,
    x='Tienda',
    y='Facturación Numerica',
    title='Facturación Total por Tienda',
    color='Tienda',
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_traces(
    textposition='none',
    marker=dict(line=dict(width=1.2, color='black'))
)

fig.update_layout(
    xaxis_title="Tienda",
    yaxis_title="Facturación (COP)",
    yaxis_tickformat=",",
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='white',
    title_font_size=20
)

fig.show()
```

# 4. Productos más y menos vendidos

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(tiendas, start=1):
    print(f"Tienda {i}:")

    conteo = df['Producto'].value_counts()

    mas_vendido = conteo.idxmax()
    ventas_max = conteo.max()

    menos_vendido = conteo.idxmin()
    ventas_min = conteo.min()

    print(f"Más vendido: {mas_vendido} ({ventas_max} ventas)")
    print(f"Menos vendido: {menos_vendido} ({ventas_min} ventas)")
    print("-" * 40)

    df_total = pd.concat(tiendas)


conteo_total = df_total['Producto'].value_counts()

print("Análisis General de Ventas:")
print(f"Producto más vendido: {conteo_total.idxmax()} ({conteo_total.max()} ventas)")
print(f"Producto menos vendido: {conteo_total.idxmin()} ({conteo_total.min()} ventas)")
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(tiendas, start=1):
    print(f" Tienda {i}:")

    conteo = df['Producto'].value_counts()

    # Top 5 más vendidos
    top5 = conteo.head(5).reset_index()
    top5.columns = ['Producto', 'Cantidad']

    fig_top = px.bar(
        top5,
        x='Cantidad',
        y='Producto',
        orientation='h',
        title=f"Tienda {i} - Top 5 Productos Más Vendidos",
        text='Cantidad',
        color='Cantidad',
        color_continuous_scale='Blues'
    )
    fig_top.update_traces(textposition='outside')
    fig_top.update_layout(yaxis={'categoryorder':'total ascending'})
    fig_top.show()

    # Top 5 menos vendidos
    low5 = conteo.tail(5).reset_index()
    low5.columns = ['Producto', 'Cantidad']

    fig_low = px.bar(
        low5,
        x='Cantidad',
        y='Producto',
        orientation='h',
        title=f"Tienda {i} - Top 5 Productos Menos Vendidos",
        text='Cantidad',
        color='Cantidad',
        color_continuous_scale='Reds'
    )
    fig_low.update_traces(textposition='outside')
    fig_low.update_layout(yaxis={'categoryorder':'total ascending'})
    fig_low.show()
```

# 5. Envío promedio por tienda

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(tiendas, start=1):
    promedio = df['Costo de envío'].mean()
    print(f'Promedio costo de envío tienda {i}: {promedio:.2f}')
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = [f"Tienda {i+1}" for i in range(len(tiendas))]

promedios_envio = [round(df['Costo de envío'].mean()) for df in tiendas]

df_envios = pd.DataFrame({
    'Tienda': nombres,
    'Promedio Costo de Envío': promedios_envio
})

max_valor = df_envios['Promedio Costo de Envío'].max()
min_valor = df_envios['Promedio Costo de Envío'].min()

def asignar_color(valor):
    if valor == max_valor:
        return 'red'
    elif valor == min_valor:
        return 'green'
    else:
        return 'lightgray'

df_envios['Color'] = df_envios['Promedio Costo de Envío'].apply(asignar_color)

fig = px.bar(
    df_envios,
    x='Tienda',
    y='Promedio Costo de Envío',
    text='Promedio Costo de Envío',
    title='Costo Promedio de Envío por Tienda (Valores Cerrados)',
    color='Color',
    color_discrete_map='identity'
)

fig.update_traces(texttemplate='$%{text}', textposition='outside')
fig.update_layout(
    yaxis_tickprefix="$",
    yaxis_title='Costo Promedio de Envío (COP)',
    showlegend=False
)
fig.show()
```

El objetivo es identificar la tienda con menor rendimiento, sustentando la recomendación en datos cuantitativos y visualizaciones obtenidas durante el análisis.

### Desarrollo

**Ingresos Totales**
La Tienda 1 lidera en ingresos acumulados durante el periodo analizado, seguida por las Tiendas 2 y 3. **Ventas por Categoría y Productos Más y Menos Vendidos**
Todas las tiendas comercializan las mismas categorías, destacando "Muebles" y "Electrónicos" como las categorías con mayor volumen de ventas en la mayoría de los casos. No obstante, la Tienda 4, a pesar de su menor facturación general, muestra un mejor desempeño relativo en la categoría "Juguetes". En cuanto a productos específicos, los más vendidos corresponden principalmente a muebles y dispositivos electrónicos, especialmente en las Tiendas 1 y 3, mientras que los menos vendidos suelen ser libros e instrumentos musicales.

**Calificaciones Promedio**
Las calificaciones promedio de las cuatro tiendas se mantienen en torno a 4 sobre 5, con poca variación. La Tienda 4 presenta calificaciones similares a las demás, a pesar de su menor volumen de ventas.

**Costo de Envío Promedio**
Existe una correlación clara entre el volumen de ventas y el costo promedio de envío. La Tienda 1, con el mayor volumen de ventas, también presenta el costo de envío promedio más alto, mientras que la Tienda 4 refleja el costo más bajo, acorde con su menor actividad comercial.
